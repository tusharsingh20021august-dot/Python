```python
# if,elif,else

age = int(input("Enter the Age: "))

if age <= 0:
    print("Invalid Age")
elif age >= 90:
    print("Greater than 60")
elif age >= 18:
    print("Eligible for Voting")
else:
    print("Error occurred")
```

```python
# program till i =5

i = 0 
while i <= 5:
    '''
    1st iteration
    i = 1, i<=5 = True

    2nd iteration
    i=2, i<=2 = True

    5th iteration
    i = 5, i<=5 = True

    6th iteration
    i = 6, i<=6 = False

    '''
    i+=1
    print(i)
```

```python
#for loop
# -  A for loop repeats a block of code for each item in a sequence (like a list, string, or range of numbers).

a = [1,2,3,4,5]

for i in a:
    '''
    i = variable
    a = sequence

    i = 1
    i = 2
    i = 5
    print(i)
    '''

    print(i)
```

```python
#Nested loops
'''
A nested loop means one loop inside another loop.
It’s useful when you need to repeat something inside another repeated process.
'''

for i in range(1, 3):        # Outer loop
    for j in range(3, 6):    # Inner loop
        print(i, j)
```

```python
# break Statement it is use to stop the loop imiditaly

for i in range(1,11):
    if i == 6:
        break
            
    print(i)
```

```python
# continue statement it is used to skip the current iteration and move to  next one...


for num in range(1,11):
    if num == 6:
        continue #skip
            
    print(num)
```

```python
#pass

'''
pass is a placeholder — it means “do nothing.”
It’s used when Python expects some code, but you don’t want to write anything (yet).
'''
for num in range(1, 6):
    if num == 3:
        pass  # do nothing for 3
    else:
        print(num)
```

```python
#print even or odd number

for i in range(1,10):
    if i %2 == 0:
        print('Even',i)
    else:
        print('Odd',i)
```
