```python
# Paraameters and arguments : it is used to it define the place holder  which calls the function 
'''
Parameters = Place holders for the values
Arguments  = 'Tushar'
'''
def greeting(name):   #Parameters
    ''' displaying a hi message to user''' #dcostring
    print('Hii',name)
greeting("tushat")   #Arguments
greeting('Amit')     #Arguments
```

```python
# 3 types of arguments :

'''
Postional arguments 
Keyword arguments
Default arguments
'''
# 1 Postional arguments

def greeting(name,city):   #Parameters
    ''' displaying a hi message to user''' #dcostring
    print('Hii',name,'welcomr to the',city)
    
greeting('Tushar','mumbai')  # we set the value on the given position
```

```python
# 2 Keyword Arguments


def greeting(name,city):   #Parameters
    ''' displaying a hi message to user''' #dcostring
    print('Hii',name,'welcomr to the',city)
    
greeting(name='vaibhav',city='banglore')  # we set the value on key data (name = 'tushar')
```

```python
#3 Default arguments


def greeting(name ='sahil',city='Lucknow'):   #already default value is set Parameters 
    ''' displaying a hi message to user''' #dcostring
    print('Hii',name,'welcomr to the',city)
    
greeting()   # default value is presented
greeting(name='vaibhav',city='banglore')  #in case we get the original data than we can make the changes
```

```python
# 4 Return : When you use return, the function sends a value back to the caller â€” instead of just displaying it on the screen.

def greeting(name, city):
    '''Returning a hi message to the user'''
    message = f'Hi {name}, welcome to {city}'
    return message

# Store the returned value in a variable
result = greeting('Tushar', 'Mumbai')
print(result)
```

```python
# 5 Decorator - In main def function we add/modifies one more function without any changes in the code...


def my_decorator(func):
    def wrapper():
        print('something is happening before the function')
        func()
        print('something is happening afterthe function')
    return(wrapper)

@my_decorator
def say_hello():
    print('hello')

say_hello()
```

```python
# 1) Print first and last name

# Function that takes two parameters: name and last name
def greet(name, last_name):
    print('Hello', name, last_name)

# Calling the function with two arguments
greet('Tushar', 'Singh')
```

```python
# 2) Add two numbers
# 2 methdos

# Method 1: Passing two numbers directly to function
def add_num(number, number1):
    print('Sum is:', number + number1)

# Call the function with two numbers
add_num(10, 15)
```

```python
# Method 2: Taking input from user
def add(total):
    print('Total number is:', total)

# Take input from user
num = int(input('Enter the first number: '))
num2 = int(input('Enter the second number: '))
# Pass the sum of two numbers to the function
add(num + num2)
```

```python
# 3) Subtraction of two numbers
# 2 methdos

# Method 1: Subtraction inside function and checking result
def subtraction(number, number1):
    result = number - number1       # Perform subtraction
    if result < 0:
        print('Result is a Negative number')
    else:
        print('Result is a Positive number')
    print('Subtraction Result:', result)

# Call the function
subtraction(30, 200)
```

```python
# Method 2: Taking input from user
def minus(num1, num2):
    result = num1 - num2
    if result < 0:
        print('Result is Negative')
    else:
        print('Result is Positive')
    print('Subtraction Result:', result)

# Take user input and call the function
number = int(input('Enter the first number: '))
number1 = int(input('Enter the second number: '))
minus(number, number1)
```

```python
# 4) Find Even or Odd number

# First Method
def Even_and_odd(num):
    # Check if number is divisible by 2
    if num % 2 == 0:
        print('Number is Even')
    else:
        print('Number is Odd')

# Take input from user and call the function
number = int(input('Enter the number: '))
Even_and_odd(number)
```
