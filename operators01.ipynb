```python
# Arthemathics operators

x = 5
y = 4

print(x+y)
print(x-y)
print(x*y)
print(x//y)
print(x%y)   
print(x**y)  #power(5*5*5*5*5 = 625)  
```

```python
# Comparision operators
x = 5
y = 3

print(x==y)    # equal to
print(x!=y)     #not equal to
print(x>y)     #greater
print(x<y)     #smaller
print(x>=y)    
print(x<=y) 
```

```python
#logical operation  it used to perform multiple operations
'''
and = need to satisfied all the condition
or = atleast one condition need to satisfied
not = reverse the statement

'''

a = 10
b = 20

print('and operator')
print(a<b and a==12)
print(a<b and a==10)
print('OR operator')
print(a<b and a==10)
print('not operator')
print(not(a<b and a==10))    # its true statement but by using of not operstor it reveser the statement from true to false
```

```python
# Assignment operations

'''

=
+=
-=
*=

'''

x = 20

x+=5
print(x)


y = 15
y-=5
print(y)

y1 = 5
y1 *=5
print(y1)
```

```python
# Identity operators 

'''

is - if the memory is on same location than only it will present (true or else false)
is not - if the memory location is differnt than only present the (true or else false)

'''

a = [1,2,3,4]
b = a
c = [2,7,8,9]

print('IS operator')
print(a is b)
print(a is c)

print('IS NOT operator')
print(a is not b)
print(a is not c)
```

```python
# membership operator is used to check the value is present or not 

"""

in = it present the value present 
in not = it present the value is not present

"""

vegetables = ['Brinjal', 'Potato','leadifinger','Onion']

print('Brinjal' in vegetables)
print('flower' not in vegetables)
```
