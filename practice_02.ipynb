```python
def percentage(Marks):
    if Marks >= 85 and Marks < 100:
        return('A')
    elif Marks >= 75 and Marks < 84:
        return('B')
    else:
        return('C')
        
percentage(100)
```

```python
df['per_%'] = df['Marks'].apply(percentage)
```

```python
df
```

```python
ascending = df.sort_values(by = ['per_%'])
print(ascending)
```

```python
practice_02 = df.sort_values(by = ['per_%'])
practice_02.to_csv('practice_02', index = False)
```
