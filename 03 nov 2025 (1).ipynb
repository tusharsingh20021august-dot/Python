```python
import pandas as pd
```

```python
stu = pd.DataFrame({
    'stu_id': [101,102,103,104,105,106,107],
    'Name' : ['Joy','Sam', 'ram', 'sham','raj', 'Nakul', 'nirov'],
    'Age': [22,21,23,24,25,26,27]
   })
```

```python
score = pd.DataFrame(
    {
         'stu_id': [101,102,103],
        'sub': ['math', 'science', 'history'],
        'Score' :[75,82,61]
        
    })
```

```python
print(stu)
print('-----------------------------------------------------')
print(score)
```

```python
# inner join (merge())

inner_merge = pd.merge(stu,score, on ='stu_id')
print(inner_merge)
```

```python
# left join

left_join = pd.merge(stu,score, on ='stu_id',how = 'left')
print(left_join)
```

```python
# Right_ join

righ_join = pd.merge(stu,score, on ='stu_id',how = 'right')
print(righ_join)
```

```python
# full_join

full_join = pd.merge(stu,score, on= 'stu_id', how = 'outer')
print(full_join)
```
