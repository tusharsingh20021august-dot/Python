```python
import matplotlib.pyplot as mt
import seaborn as sns
import pandas as pd
```

```python
# load dataset (upload through online)

tips = sns.load_dataset('tips')
print(tips.head())
```

```python
tips.info()
```

```python
tips.describe()
```

```python
# scatterplot 

sns.scatterplot(data = tips,  x='total_bill', y='tip', hue = 'time')

# set tittle
mt.title('total_bill vs  tip')   # it present the title 
mt.xlabel('total_bill')    # it present x axis
mt.ylabel('tip')            # it present y axis
```

```python
# scatterplot 

sns.scatterplot(data = tips,  x='total_bill', y='tip', hue = 'time', style = 'smoker')
```

```python
# line plot
sns.lineplot(data = tips,  x='total_bill', y='tip', hue = 'time')

```

```python
# load new dataset

fmri =sns.load_dataset('fmri')
print(fmri.head())
```

```python
# lineplot

sns.lineplot(data=fmri,x='timepoint',y='signal',hue='event',style='region')
mt.title('FMRI signal over time') 
mt.xlabel('timepoint')
mt.ylabel('signal')
mt.show()
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
# barplot

sns.barplot(data =tips, x='day', y='total_bill', hue ='time')

# title

mt.title('Average total bill per Day by time')
mt.xlabel('day')
mt.ylabel('Avg toal bill')
mt.savefig('Barplot.png')       # used to save the data
```

```python
# histogram total_bill - distribution  of  total_bill

sns.histplot(data = tips, x='total_bill', bins = 10, kde=True)
# kde - kernel density estimation

mt.title('Distribution if total_bill')
mt.xlabel('Total_bill')
mt.ylabel('Frequency')
```

```python
# boxplot - summarize data distribution with respect to categories
# highligh outliers

sns.boxplot(data =tips, x='day', y='total_bill', hue ='smoker')
mt.title('Distribution if smoker')
mt.xlabel('day')
mt.ylabel('time')
```

```python
# load dataset

flower = sns.load_dataset('iris')
print(flower.head())
```

```python
# pairplot

sns.pairplot(data =flower, hue='species')

mt.suptitle('Iris Pairplot', y = 1.03)   # y is used to allocate the space 
```

```python
# violinplot

sns.violinplot(data =flower,x='species', y='sepal_length', hue='species')

```

```python
# coutplot

sns.countplot(data =flower,x='species')
mt.title('count Iris species')
mt.xlabel('species')
mt.ylabel('count')
```

```python
# facet frid = multiple plots based on categories

g = sns.FacetGrid(data =tips,col='time', row='smoker')
g.map(sns.scatterplot,'total_bill', 'tip')
g.add_legend()
```

```python
# joint plot = combination of scatter plot and histogram

sns.jointplot(data = tips, x='total_bill', y='tip', kind='scatter',hue='time')
mt.suptitle('Joint plot of total')
```
