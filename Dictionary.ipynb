```python
# Dictionary {} (crule bruces) 

# Intro:-
# in dic the data is store in pair(key : value) order form
# the dic are unorderd
# It dynamic 

dic = {
    'name' :  'Tushar',       # name - Key  :  ushar - value   [ in key we can list[], tuple() can also store in the dic)
    'Age' : 26,               # key - its is always unquie
    'Marks' : 55
}
print(dic)
```

```python
# 1) Add 

d1 = {
    'Fruits' : ['Apple', 'banana' , 'orange '],
    'categoey' : 'fruits'
     }
print(d1)

d1['price'] = 100
print(d1)
```

```python
# 2) Update : how update the existing 

dic = {
    'name' : [ 'Tushar', 'Harrry', 'sahil'  , 'brijes' ] ,              
    'Marks' :[ 55,65,82,69]
}
print(dic)


dic['Div'] = ['A']           # Add the new Key(Div) in the dic
print(dic)

dic['Div'] = ['B']    # update the div : from "A" to "B"
print(dic)

```

```python
# Delete : this help to delete the data in the ditionary

del dic['Div']
print(dic)
```

```python
# comprehension - It’s a short and elegant way to create a new list from an existing one — all in one line of code.

square = {x: x**x for x in range(1,6)}
print(square)
'''
Explanation:
Here x is a variable, not a string.
For each x in range(1,6) → keys are 1, 2, 3, 4, 5
Values are x**x (power of itself)Explanation:

'''

square = {'x': x**x for x in range(1,6)}
print(square)
'''
Explanation:
Here 'x' is a string, not a variable.
That means every key is literally 'x'.[5*5*5*5*5 = 3125]
Since all keys are the same, each new value overwrites the previous one.
'''
```
