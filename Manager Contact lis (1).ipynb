```python
print("welcome to Manager Contact list")

contacts = {}   # Dictionary for storing contacts
phone_set = set()   # Prevent duplicate phone numbers

def add_contact():
    """Add a new contact"""
    name = input("Enter name: ")
    phone = input("Enter phone: ")
    email = input("Enter email: ")

    if phone in phone_set:
        print(" Phone number already exists!")
        return

    contacts[name] = (phone, email)   #  tuple
    phone_set.add(phone)
    print("Contact added successfully")

def update_contact():
    """Update an existing contact"""
    name = input("Enter name to update: ")
    if name not in contacts:
        print(" Contact not found.")
        return

    phone = input("Enter new phone: ")
    email = input("Enter new email: ")

    if phone in phone_set and phone != contacts[name][0]:
        print(" Phone number already used by another contact.")
        return

# ---------------- Main Program ----------------
while True:
    print(" Contact Book Manager ")
    print("1. Add Contact")
    print("2. Update Contact")
    print("3. Exit")

    choice = input("Enter choice (1-3): ")

    if choice == "1":
        add_contact()
    elif choice == "2":
        update_contact()
    elif choice == "3":
        print(" Exiting Contact Book. Goodbye!")
        break
    else:
        print(" Invalid choice, try again.")
```
