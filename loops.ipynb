```python
def student_info(*args,**kwargs):
    print(args)       # tuple of all positional arguments      
    print(kwargs)     # dict of all keyword arguments

courses = ['mahts','science']                         #courses is a list.
info    = {'name': 'tushar','age' : 22,'Div':'B'}     #info is a dictionary.

student_info(*courses,**info)

#courses → unpacks the list into separate positional arguments:
#- mahts', 'science'
#info → unpacks the dictionary into keyword arguments:
#- name='tushar', age=22#
    
    
```

```python
# Return 'Even' if the number is even, otherwise 'Odd'.
def even_or_odd(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"
```

```python
print(even_or_odd(6))
```
