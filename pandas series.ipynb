```python
#instead print used return than it will be display the value
def cat(x):
    if(x>300):
        return('High Range')
    elif(x<=299) and (x>=150):
        return('Medium Range')
    else:
        return('low Range')

cat(100)
```

```python
rd_1['Range'] =rd_1['total_exp'].apply(cat)
```
