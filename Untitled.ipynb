```python
# common elemwenta - in this we find the common values

a = [1,2,3,4,5,6,55,11,22]
b = [22,45,55,6,5,4,1]

#set function       # convert into sets
s1 = set(a)
s2 = set(b)

s3 = s1.intersection(s2)    # we intersect the s1 and s2
print(list(s3))             #this list we use to conver the set from list
```
